#### 11.Reverse a number.

```python
num = int(input("Enter a number :"))
rev = 0
while (num > 0):
    digital = num % 10        #    123 % 10 = 3
    rev = rev * 10 + digital
    num = num // 10            #    123 // 10 = 12
print(f"Reverse number :",rev)
```

#### 15.Check if a number is prime.

```python
num = int(input("Enter a number : "))
is_prime = True
if num < 2:
    is_prime = False
else:
    for i in range(2, int(num ** 0.5)+1):
        if num % 2 == 0:
            is_prime = False
            break
    if is_prime:
        print(f"{num} is a Prime number")
    else:
        print(f"{num} is not a Prime number")
```

#### 17.Check if a number is a palindrome.

#### A palindrome is a word, number, or sentence that reads the same forward and backward.

```python
num = int(input("Enter a number : "))
temp = num
rev = 0
while temp > 0:
    rev = rev * 10 + temp % 10
    temp //= 10
if num == rev:
    print(f"{num} is a Palindrome number.")
else:
    print(f"{num} is not a Palindrome number.")
```

#### 18.Check if a number is an Armstrong number.

#### An Armstrong number is a number where sum of each digit raised to the power of total digits = original number

```python
num = int(input("Enter a number : "))
temp = num 
order = len(str(num))
sum = 0
while temp>0:
    digital = temp%10
    sum+= digital ** order
    temp //= 10
if sum==num:
    print(f"{num} is Armstrong number")
else:
    print(f"{num} is not Armstrong number")
```

#### 24.Find GCD (Greatest Common Divisor) of two number.

#### 25.Find LCM (Least Common Multiple) of two number.

```python
a, b = map(int, input("Enter two number : ").split())
max_num = max(a,b)
while True:
    if max_num%a == 0 and max_num%b == 0:
        print("LCM = ",max_num)
        break
    max_num+=1
```

#### 26.Count even and odd digits in a number.

```python
num = int(input("Enter a number : "))
even = odd = 0
while num > 0:
    digit = num % 10
    if digit % 2 == 0:
        even += 1
    else:
        odd += 1
    num //= 10
print("Even digits = ",even)
print("Odd digits = ",odd)
```
