## Input & Output in Python 

#### Output :- The print() function is used to display output on the screen.

```python
# Basic print 

print("Hello World!")
```

```python
# Printing Multiple value

name = "Alice"
age = 17
print("Name :" , name ," , " ,"Age :" , age)
```

```python
# Using sep & end parameter

print("Hello","World!",sep= "-")
print("Python",end = " ")      # new line
print("is awesome!")
```

```python
# Formate output (using f-string)

name = "Alice"
age = 25
print(f"My name is {name} and I am {age} years old!")
```

#### Input :- The input() function is used to take input from the keyboard. It always returns a string.

```python
# Basic Input

name = input ("Enter your name :")
print("Hello," + name + "!")
```

```python
# Taking Input

age = int(input("Enter your age : "))
print(f"Your are {age} years old!")
```

```python
# Taking Multiple Input

x, y = map(int,input("Enter rwo numbaers:").split())
print(f"Sum : {x+y} ")
```

### Mutli words varibale Names

#### Camel-case :- Each word,except the first , starts with a capital letters.

```python
myVariableName = 'John'
myVariableName
```

#### Pascal-case :- Each word starts with a capital letter.

```python
MyVariableName = 'john'
MyVariableName
```

#### Snake-case :- Each word is separated by an underscore character.

```python
my_Variable_name = 'John'
my_Variable_name
```

### Assign Mutliple value

```python
# Many value to mutliple variable

x,y,z = 'orange','Banana','Cherry'
print(x)
print(y)
print(z)
```

```python
# One value to mutliple variable

x = y = z = "Orange"
print(x)
print(y)
print(z)
```

#### Swaping two variable 

```python
a,b = 5, 2
print(a)
print(b)
print('Swaping')
a,b = b,a
print(a)
print(b)
```
