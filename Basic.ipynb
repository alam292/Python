## Input & Output in Python 

#### Output :- The print() function is used to display output on the screen.

```python
# Basic print 

print("Hello World!")
```

```python
# Printing Multiple value

name = "Alice"
age = 17
print("Name :" , name ," , " ,"Age :" , age)
```

```python
# Using sep & end parameter

print("Hello","World!",sep= "-")
print("Python",end = " ")      # new line
print("is awesome!")
```

```python
# Formate output (using f-string)

name = "Alice"
age = 25
print(f"My name is {name} and I am {age} years old!")
```

#### Input :- The input() function is used to take input from the keyboard. It always returns a string.

```python
# Basic Input

name = input ("Enter your name :")
print("Hello," + name + "!")
```

```python
# Taking Input

age = int(input("Enter your age : "))
print(f"Your are {age} years old!")
```

```python
# Taking Multiple Input

x, y = map(int,input("Enter rwo numbaers:").split())
print(f"Sum : {x+y} ")
```
