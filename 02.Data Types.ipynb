### Integer (int)

#### An integer is a numeric data type that represents whole numbers without decimals.It can be positive, negative, or zero.

#### Properties of Integer

Data type: int

Immutable

No size limit (depends on memory)

Used for counting and calculations

```python
a = 85
a
```

```python
b = -55
b
```

```python
c = 0
c
```

```python
type(a)
```

```python
print(type(9))
```

### Float (float)

#### A float is a numeric data type that represents numbers with decimal points or fractional values.

#### Properties of Float

Data type: float

Immutable

Stores decimal values

More memory than int

Can have precision limitations

```python
a = 4.5
a
```

```python
print(type(a))
```

```python
b = 123E34
b
```

```python
type(b)
```

### Complex (complex)

#### A complex number is a number that has two parts: Real part & Imaginary part

In Python, it is written as:

ùëé+ùëèùëó

(where j represents ‚àö‚àí1)

#### Properties

Data type: complex

Immutable

Uses j instead of i

Supports mathematical operations

```python
a = 3+8j
a
```

```python
type(a)
```

```python
z = 3 + 4j
print(z.real)   # 3.0
print(z.imag)   # 4.0
```

```python
print(type(z))
```

```python
a = 2 + 3j
b = 1 + 2j

print(a + b)
print(a * b)
```

#### Random Number

Python generates random numbers using the random module.

```python
import random
print(random.randint(1,10))
```

```python
type(random.randint(1,10))
```

```python
# üîπ Random Float (0 to 1)

print(random.random())
```

```python
# üîπ Random Float in a Range

print(random.uniform(5, 10))
```

```python
# üîπ Random Choice from List

colors = ["red", "blue", "green"]
print(random.choice(colors))
```

```python
# üîπ Random Shuffle

nums = [1, 2, 3, 4, 5]
random.shuffle(nums)
print(nums)
```
