# Type Conversion 

#### Type conversion is the process of converting one data type into another using built-in functions.

### 1.Implicit Type Conversion (Automatic)

Python converts data types automatically.

✔ int → float automatically

```python
a = 10      # int
b = 2.5     # float

c = a + b
print(c)        # 12.5
print(type(c))  # float
```

### 2.Explicit Type Conversion (Type Casting)

Explicit type conversion means the programmer manually converts one data type into another using built-in functions.

Why do we use Explicit Type Conversion?

✔ To avoid type errors

✔ To take user input correctly

✔ To control data processing


```python
x = "25"
y = int(x)
print(y, type(y))
```

```python
a = 10
b = float(a)
print(b)
```

```python
num = 500
s = str(num)
print(s)
```

```python
lst = [1, 2, 2, 3]
s = set(lst)
print(s)
```
